# 5. Voiced stops (bvot, dvot, gvot)

```python
# SIMULATING VOICED (bdg) VALUES FROM VOICELESS (ptk)
# SHOULD NOT BE PART OF THE NOTEBOOK
Fa17['bvot'] = Fa17.pvot + 0.05
Fa17['dvot'] = Fa17.tvot + 0.05
Fa17['gvot'] = Fa17.kvot + 0.05
```

We will begin by adding a column for average voiced and average voiceless for each person in our table:

```python
Fa17['avg voiceless (ptk)'] = Fa17[['pvot', 'tvot', 'kvot']].mean(numeric_only=True, axis=1)
Fa17['avg voiced (bdg)'] = Fa17[['bvot', 'dvot', 'gvot']].mean(numeric_only=True, axis=1)
Fa17.head()
```

Now, we will compare our data.

```python
ax = sns.lmplot('avg voiceless (ptk)', 'avg voiced (bdg)', data=Fa17, fit_reg=False)
```

```python
ax = sns.lmplot('avg voiceless (ptk)', 'height', data=Fa17, fit_reg=False)
```

```python
ax = sns.lmplot('avg voiced (bdg)', 'height', data=Fa17, fit_reg=False)
```
